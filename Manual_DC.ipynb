```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
'/content/drive/MyDrive/cafe_sales_dirty.csv'
```

```python
# Load
df = pd.read_csv('/content/drive/MyDrive/cafe_sales_dirty.csv')
df.head()
```

```python
# Check Datatype
df.dtypes
```

```python
# Convert to numeric
# df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Quantity'] = df['Quantity'].fillna(0).astype(int)
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
df.dtypes
```

```python
# Find Missing values
df.isnull().sum()
```

```python
# Filling Missing values
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].mean())
df['Total Spent'] = df['Total Spent'].fillna(df['Total Spent'].mean())
```

```python
df.isnull().sum()
```

```python
# Replacing ERROR with Unknown
df.replace(['ERROR','Unknown'], pd.NA, inplace=True)
```

```python
# Filling values into items
mode_item = df['Item'].mode()[0]
df['Item'] = df['Item'].replace('Unknown' and 'UNKNOWN', np.nan)
df['Item'] = df['Item'].fillna(mode_item)
```

```python
# Filling values into Payment Method
mode_pm = df['Payment Method'].mode()[0]
df['Payment Method'] = df['Payment Method'].replace('Unknown' and 'UNKNOWN', np.nan)
df['Payment Method'] = df['Payment Method'].fillna(mode_pm)
```

```python
# Filling missing values in location
mode_location = df['Location'].mode()[0]
df['Location'] = df['Location'].replace('Unknown' and 'UNKNOWN', np.nan)
df['Location'] = df['Location'].fillna(mode_location)
```

```python
df.head()
```

```python

# Checking Null values
df.isnull().sum()
```

```python
# Why meadian is used in Transaction date
dat = df['Transaction Date'].median()
df['Transaction Date'] = df['Transaction Date'].fillna(dat)
df.isnull().sum()
```

```python
# Check Duplicate Values
df.duplicated().sum()
```

```python
# Removing duplicate values
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
#
df.info()
```

```python
df.to_csv('new_cafe_sales_data.csv', index=False)
print("Data Cleaning completed!")
```

# The 10 Business Questions & Answers

Q.1 Which item generated the highest revenue?

```python
df.groupby('Item')['Total Spent'].sum().sort_values(ascending=False).head(1)
```

Q.2 Which location generated the most sales?

```python
df.groupby('Location')['Total Spent'].sum().sort_values(ascending=False)
```

Q.3 Total revenue

```python
df['Total Spent'].sum()
```

Q.4 Most used payment method

```python
df['Payment Method'].value_counts()
```

Q.5 Top 5 selling items

```python
df['Item'].value_counts().head()
```

Q.6 Highest sales day

```python
df.groupby('Transaction Date')['Total Spent'].sum().sort_values(ascending=False).head(1)
```

Q.7 Average spending per transaction

```python
df['Total Spent'].mean()
```

Q.8 Item with highest average quantity sold

```python
df.groupby('Item')['Quantity'].mean().sort_values(ascending=False)
```

Q.9 Sales contribution by location (%)

```python
location_sales = df.groupby('Location')['Total Spent'].sum()
(location_sales / location_sales.sum()) * 100
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df.groupby(df['Transaction Date'].dt.month)['Total Spent'].sum()
```
