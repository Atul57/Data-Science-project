```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/ecommerce_metrics_small.csv')
df.head()
```

```python
# Line Plot
plt.figure(figsize=(8, 6))
plt.plot(df['Primary_Region'], df['Daily_Revenue_USD'], marker='o')
plt.title('Region Over Daily Revenue')
plt.xlabel('Primary_Region')
plt.ylabel('Daily_Revenue_USD')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,9))

plt.plot(df['Primary_Region'],df['Ad_Spend_USD'],marker='o')
plt.plot(df['Primary_Region'],df['Active_Users'],marker='o')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter(
    df['Ad_Spend_USD'],
    df['App_Downloads'],
    color='tab:orange',
    alpha=0.7,
    edgecolors='black'
)
plt.title('Relationship between Ad Spend and App Downloads', fontsize=14)
plt.xlabel('Ad Spend (USD)', fontsize=12)
plt.ylabel('App Downloads', fontsize=12)
plt.show()
```

```python
region_counts = df['Primary_Region'].value_counts()
plt.pie(region_counts, labels=region_counts.index)
plt.show()
```

```python
revenue_by_region = df.groupby('Primary_Region')['Daily_Revenue_USD'].sum()
plt.figure(figsize=(8, 5))
plt.bar(revenue_by_region.index, revenue_by_region.values, color='tab:blue', edgecolor='black')

# 4. Add titles and labels
plt.title('Total Revenue by Region (30 Days)', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Revenue (USD)', fontsize=12)
plt.show()
```
