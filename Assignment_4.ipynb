Q . 6 = Filter students where Score > 80. How many students scored above 80?

```python
students_above_80 = df[df['Score'] > 80]
display(students_above_80)
num_students_above_80 = len(students_above_80)
print(f"Number of students who scored above 80: {num_students_above_80}")
```

Q . 7 = Filter students from Delhi
who are in CSE
department. Show Name,
Score, and Grade only.


```python
filtered_students = df[(df['City'] == 'Delhi') & (df['Department'] == 'CSE')]
display(filtered_students[['Name', 'Score', 'Grade']])
```

Q . 8 = Find all students who
Failed (Passed == False).
What is their average
Score?

```python
failed_students = df[df['Passed'] == False]
display(failed_students)

average_score_failed = failed_students['Score'].mean()
print(f"Average score of students who failed: {average_score_failed:.2f}")
```

Q . 9 = Filter students whose
Name contains the string
'a' (case-insensitive) using
.str.contains().

```python
students_with_a_in_name = df[df['Name'].str.contains('a', case=False)]
display(students_with_a_in_name)
```

Q . 10 = Add a new column
'Score_Category': if Score
>= 80 → 'Distinction', Score
>= 60 → 'Merit', else
'Pass/Fail'. Use np.where()
or nested condition.


```python
df['Score_Category'] = np.where(df['Score'] >= 80, 'Distinction',
                                  np.where(df['Score'] >= 60, 'Merit', 'Pass/Fail'))
display(df)
```

Sort the DataFrame by
Score in descending order.
Reset the index. Show the
top 5 students.

```python
df_sorted = df.sort_values(by='Score', ascending=False).reset_index(drop=True)
display(df_sorted.head(5))
```
